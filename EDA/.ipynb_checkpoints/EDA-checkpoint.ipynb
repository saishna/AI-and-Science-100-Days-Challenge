Corelation 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Creating the DataFrame
data = {
    'Age': [22, 25, 28, 30, 35, 40, 45],
    'Income': [25000, 35000, 45000, 50000, 65000, 70000, 80000]
}
df = pd.DataFrame(data)
```

```python
# Calculating the Pearson correlation coefficient
correlation = df.corr()
print(correlation)
```

```python
# Plotting the scatterplot
sns.scatterplot(x='Age', y='Income', data=df)
plt.title('Scatterplot of Age vs. Income')
plt.xlabel('Age')
plt.ylabel('Income ($)')
plt.show()
```

```python
# Adding a new variable for Education Level
df['Education_Level'] = [1, 2, 2, 3, 3, 3, 3]

# Calculating the correlation matrix
correlation_matrix = df.corr()
print(correlation_matrix)

# Visualizing the correlation matrix using a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

Exploring TWo or more variavbles

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(0)

# Simulated data for "Finished Square Feet" and "Tax-Assessed Value"
sqft = np.random.normal(loc=2000, scale=500, size=1000)  # Finished square feet
tax_value = sqft * 300 + np.random.normal(loc=0, scale=20000, size=1000)  # Tax-assessed value with some noise

# Creating DataFrame
data = pd.DataFrame({
    'FinishedSquareFeet': sqft,
    'TaxAssessedValue': tax_value
})
```

```python
# Hexagonal bin plot
plt.figure(figsize=(8, 6))
plt.hexbin(data['FinishedSquareFeet'], data['TaxAssessedValue'], gridsize=30, cmap='Blues')
plt.colorbar(label='Density')
plt.xlabel('Finished Square Feet')
plt.ylabel('Tax-Assessed Value')
plt.title('Hexagonal Binning of Finished Square Feet vs. Tax-Assessed Value')
plt.show()
```

```python
# Contour plot with Seaborn
plt.figure(figsize=(8, 6))
sns.kdeplot(
    data=data,
    x='FinishedSquareFeet',
    y='TaxAssessedValue',
    cmap='viridis',
    fill=True,
    thresh=0,
    levels=10
)
plt.xlabel('Finished Square Feet')
plt.ylabel('Tax-Assessed Value')
plt.title('Contour Plot of Finished Square Feet vs. Tax-Assessed Value')
plt.show()
```

Contigency Table for two categorial Variations

```python
import pandas as pd

# Sample data: loan grades and loan outcomes
data = {
    'grade': ['A', 'A', 'B', 'C', 'D', 'A', 'B', 'C', 'D', 'G', 'F', 'E', 'A', 'G', 'F', 'E'],
    'status': ['Fully Paid', 'Current', 'Charged Off', 'Late', 'Current', 'Late', 'Fully Paid', 'Current', 'Late', 'Charged Off', 'Fully Paid', 'Current', 'Charged Off', 'Fully Paid', 'Current', 'Late']
}

# Convert to DataFrame
loans = pd.DataFrame(data)

# Create contingency table with counts using len as a lambda function
contingency_table = loans.pivot_table(index='grade', columns='status', aggfunc=lambda x: len(x), fill_value=0, margins=True)

# Calculate row percentages
percentage_table = contingency_table.loc['A':'G',:].copy()
percentage_table.loc[:, 'Charged Off':'Late'] = percentage_table.loc[:, 'Charged Off':'Late'].div(contingency_table['All'], axis=0)

print("Contingency Table with Counts:\n", contingency_table)
print("\nContingency Table with Row Percentages:\n", percentage_table)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data of airlines and the percentage of delayed flights
data = {
    'airline': ['Alaska', 'Alaska', 'American', 'American', 'Delta', 'Delta', 'United', 'United'],
    'pct_carrier_delay': [5, 8, 40, 35, 12, 9, 25, 20]
}

# Creating a larger dataset by repeating data
np.random.seed(0)
airline_stats = pd.DataFrame({
    'airline': np.random.choice(['Alaska', 'American', 'Delta', 'United'], 200),
    'pct_carrier_delay': np.random.normal(20, 10, 200).clip(0, 50)
})

print(airline_stats.head())  # Display first few rows
```

```python
# Boxplot
plt.figure(figsize=(10, 6))
ax = airline_stats.boxplot(by='airline', column='pct_carrier_delay')
ax.set_title('Boxplot of % Carrier Delay by Airline')
ax.set_xlabel('Airline')
ax.set_ylabel('Daily % of Delayed Flights')
plt.suptitle('')  # Remove the automatic title
plt.show()
```

```python
# Violin plot
plt.figure(figsize=(10, 6))
ax = sns.violinplot(x='airline', y='pct_carrier_delay', data=airline_stats, inner='quartile', color='skyblue')
ax.set_title('Violin Plot of % Carrier Delay by Airline')
ax.set_xlabel('Airline')
ax.set_ylabel('Daily % of Delayed Flights')
plt.show()
```
